```r
dataset1<- read.delim("E:\\Restaurant_Reviews.tsv",quote = '', stringsAsFactors = FALSE)
#A tab-separated values file is a simple text format for storing data in a tabular structure
```

```r
library(tm)#cleaning text
library(SnowballC)#text stemming require the package ‘SnowballC’.
# Corpus is a list of a document
```

```r
Corpus = VCorpus(VectorSource(dataset1$Review))
Corpus
```

```r
Corpus = tm_map(Corpus, content_transformer(tolower))
as.character(Corpus[[1]])
```

```r
Corpus = tm_map(Corpus, removeNumbers)

as.character(Corpus[[1]])
```

```r

Corpus = tm_map(Corpus, removePunctuation)
as.character(Corpus[[1]])
```

```r
#specify your stopwords as a character vectorlike: this ,we,was,and,the,not etc
Corpus = tm_map(Corpus, removeWords,stopwords())
as.character(Corpus[[1]])
```

```r
#reduces words to their root form.
Corpus = tm_map(Corpus, stemDocument)
as.character(Corpus[[1]])
```

```r
# Eliminate extra white spaces
Corpus = tm_map(Corpus, stripWhitespace)
as.character(Corpus[[1]])
```

```r
dtm = DocumentTermMatrix(Corpus)
dtm
```

```r
dtm = removeSparseTerms(dtm ,0.999)# remove non frequentwords
dtm
```

```r
m <- as.matrix(dtm)
m
```

```r
v <- sort(colSums(m),decreasing=TRUE)
v
```

```r
d <- data.frame(Review = names(v),Liked=v)
d
```

```r
head(d, 10)
library(wordcloud)
wordcloud(words = d$Review, freq = d$Liked, min.freq = 1,
          max.words=200, random.order=FALSE, rot.per=0.35, 
          colors=brewer.pal(8, "Dark2"))
#words : the words to be plotted
#freq : their frequencies
#min.freq : words with frequency below min.freq will not be plotted
#max.words : maximum number of words to be plotted
#random.order : plot words in random order. If false, they will be plotted in decreasing frequency
#rot.per : proportion words with 90 degree rotation (vertical text)
#colors : color words from least to most frequent. Use, for example, colors =“black” for single color.
```
