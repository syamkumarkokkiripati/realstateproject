```python
# Data Manupulation
import numpy as np
import pandas as pd
# Plotting graphs
import matplotlib.pyplot as plt

# CSV File - Read
data=pd.read_csv("E:\\credit\\LGD.csv")
#  ,header=None ,skiprows=1 , names=['Cal', 'Pr', 'Fat', 'sod', 'Fib', 'Rting']
print(data.head(5)) 
print(data.ndim)
print(data.shape)
```

```python
# pd.get_dummies allows to convert a categorical variable into dummy variables
# Create a set of dummy variables from the sex variable
data_Gender = pd.get_dummies(data['Gender'])
data_Married = pd.get_dummies(data['Married'])
# Join the dummy variables to the main dataframe
data_new = pd.concat([data, data_Gender], axis=1)
data_new = pd.concat([data_new, data_Married], axis=1)         
data_new.head()
```

```python
dataClean = data_new[['Age','Number of Vehicles','F','Single','Losses in Thousands']].copy()
print(dataClean.head(5))
```

```python
dataClean.corr()
```

```python
dataClean = dataClean.rename(columns={'Number of Vehicles': 'nVeh', 'Losses in Thousands': 'LTh'})
print(dataClean.head(5)) 

import numpy as np
import statsmodels.api as sm 
import statsmodels.formula.api as smf
results = smf.ols('LTh ~ Age + nVeh + F + Single', data=dataClean).fit()
print(results.summary())
```
